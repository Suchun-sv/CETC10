```python
import pandas as pd
import numpy as np
import os
```

```python
class UPLOAD:
    def __init__(self, rootName=''):
        self.root = rootName
        
    def dian(self, df, name):
        df['JD'] = df['JD'].astype(str)
        df['WD'] = df['WD'].astype(str)
        df['new'] = df['JD']+','+df['WD']
        df.to_csv(name+'.csv', index=False)
        return df
    
    def oldTonew(self, df_temp):
        h = list()
        for j in df_temp:
            j = '['+j+']'
            h.append(j)
        h = str(h)
        h = h.replace("'","")
        h = h[1:-1]
        return h
    
    def xian(self, df, colName='0'):
        '''将同一组的JD和WD变成一列
        
        df是df,colName是要换的名字'''
        df['JD'] = df['JD'].astype(str)
        df['WD'] = df['WD'].astype(str)
        df['new'] = df['JD']+','+df['WD']
        df_temp = df['new']
        m = list()
        h = list()
        for j in df_temp:
            j = '['+j+']'
            h.append(j)
        h = str(h)
        h = h.replace("'","")
        h = h[1:-1]
        m = [[colName, h]]
        m = pd.DataFrame(m)
        return m
    
    def save(self, df, name):
        '''存为csv
        
        目录存在self.root里,初始化的时候确定,name就是要保存的文件名字'''
        df.to_csv(os.path.join(self.root, name+'.csv'), index=False)
```
